# Vectorized Operations
- Vectorized operations are operations that are applied simultaneously on all elements of a NumPy array without using explicit Python loops.

- Instead of processing one element at a time, NumPy processes the entire array in one operation.


# Why Vectorization Exists
- Python loops are:

   - Slow

   - Memory inefficient

   - Not suitable for large data

- NumPy vectorization:

   - Uses optimized C code internally

   - Eliminates Python-level loops

   - Executes much faster

# Comparison: Loop vs Vectorized Operation
- Using Python Loop

```python
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

result = []
for i in range(len(a)):
    result.append(a[i] + b[i])

print(result)
```

- Using NumPy Vectorization 

```python
import numpy as np

a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

print(a + b)
```

- Why Vectorized is Better

  - Shorter code

  - Faster execution

  - Cleaner syntax

# 1. Element-wise Operations
- Each element in one array interacts with the corresponding element in another array.

```python
a = np.array([2, 4, 6])
b = np.array([1, 3, 5])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

- Important Rule

   - Arrays must have:
   
      - Same shape or
      - Compatible shapes (via broadcasting)

# 2. Arithmetic Operations
- NumPy supports direct arithmetic operators.
- Operators Supported
| Operator | Meaning        |
| -------- | -------------- |
| +        | Addition       |
| -        | Subtraction    |
| *        | Multiplication |
| /        | Division       |
| //       | Floor division |
| %        | Modulus        |
| **       | Power          |


```python
a = np.array([1, 2, 3])

print(a + 5)
print(a * 2)
print(a ** 2)
```

# 3. Comparison Operations
- Comparison operations compare elements and return Boolean values (True/False).

| Operator | Description           |

| >=       | Greater than or equal |

# 4. Logical Operations
- Logical operations are used to combine multiple conditions
- Logical Operators

```python
a = np.array([10, 20, 30, 40, 50])

print((a > 20) & (a < 50))
```

- Using NumPy Functions

```python
np.logical_and(a > 20, a < 50)
np.logical_or(a < 20, a > 40)
np.logical_not(a > 30)
```

# 5. Broadcasting
- Broadcasting allows NumPy to perform operations on arrays of different shapes by automatically expanding the smaller array.

```python
a = np.array([1,2,3])
print(a + 10)
```

- How Broadcasting Works

   - Scalar 10 is stretched to [10,10,10]

   - Operation performed element-wise

- Broadcasting Rules (NO SHORTCUT)
   - Rule 1: Compare Shapes from Right
       - Compare array dimensions from right to left.



   - Rule 2: Dimensions are Compatible If
       - They are equal, OR
       - One of them is 1



   - Rule 3: Incompatible Shapes → Error
       - ValueError: operands could not be broadcast together

```python
# 1D + 2D Example
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a + b)
```

```python
# Column Broadcasting
c = np.array([[1],
              [2]])

print(a + c)
```

- Why Broadcasting is Powerful

   - Eliminates loops

   - Saves memory

   - Speeds up computation

   - Essential in ML algorithms
