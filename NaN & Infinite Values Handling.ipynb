# NaN & Infinite Values Handling in NumPy

- Real-world data is never perfect.
 - It often contains:

    - Missing values

    - Undefined values

    - Infinite values due to division errors

- NumPy provides special values and functions to handle these situations safely.

# What Are NaN and Infinite Values?
**NaN (Not a Number)**
- Represents missing or undefined numeric data
- Common in datasets after:
   - Division by zero
   - Invalid mathematical operations
   - Missing entries in files

**Infinity (inf and -inf)**
- Represents numbers larger than any finite number
- Occurs due to:
   - Division by zero
   - Overflow in calculations

## 1. np.nan
- np.nan is a special floating-point value used to represent missing data.

```python
import numpy as np

arr = np.array([10, 20, np.nan, 40])
print(arr)
```

**Important Property of NaN**
- print(np.nan == np.nan)
- Output:
    - False
- NaN never equals itself

```python
print(np.nan == np.nan)
```

# Why This Matters
- Normal comparisons do not work
- Must use special NumPy functions

## 2. np.isna() / np.isnan()
- Purpose
   - Detects NaN values in an array.
   - Both functions behave the same in NumPy.
<br><br>
- Syntax
  - np.isnan(array)
  - np.isna(array)

```python
arr = np.array([5, np.nan, 15, np.nan])
print(np.isnan(arr))
```

- Returns a Boolean mask
- True where NaN exists

```python
# Filtering NaN Values
clean = arr[~np.isnan(arr)]
print(clean)
```

## 3. np.isinf()
- Purpose
   - Detects positive and negative infinity values.
<br><br>
- Syntax
   - np.isinf(array)

```python
arr = np.array([10, np.inf, -np.inf, 20])
print(np.isinf(arr))
```

- Handling Infinite Values

```python
arr[np.isinf(arr)] = 0
print(arr)
```

## 4. np.nan_to_num()
- Purpose
   - Replaces:
     - NaN
     - +inf
     - -inf
       
with finite numbers.

**Default Behavior**
| Value | Replaced With     |
| ----- | ----------------- |
| NaN   | 0                 |
| +inf  | Very large number |
| -inf  | Very small number |

- Syntax
   - np.nan_to_num(array, nan=0, posinf=1000, neginf=-1000)

```python
arr = np.array([10, np.nan, np.inf, -np.inf])

clean = np.nan_to_num(arr, nan=0, posinf=999, neginf=-999)
print(clean)
```

**Why nan_to_num() Is Important**
- Prevents ML model crashes
- Keeps array shape unchanged
- Fast and vectorized

## 5. np.nanmean()
- Computes the mean while ignoring NaN values.

- Syntax
   - np.nanmean(array)

```python
arr = np.array([10, 20, np.nan, 40])

print(np.mean(arr))
print(np.nanmean(arr))
```

```python
# Axis Usage
arr2 = np.array([[1, np.nan, 3],
                 [4, 5, np.nan]])

print(np.nanmean(arr2, axis=0))
print(np.nanmean(arr2, axis=1))
```

## 6. np.nanstd()
- Computes standard deviation ignoring NaN values.

- Syntax
  - np.nanstd(array)

```python
arr = np.array([10, 12, np.nan, 18])

print(np.std(arr))
print(np.nanstd(arr))
```

- Why Normal std() Fails
  - Any NaN → entire result becomes NaN
