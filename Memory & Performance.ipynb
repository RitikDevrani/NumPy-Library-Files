# Memory & Performance in NumPy

NumPy is designed for high-performance numerical computing.
Its speed advantage comes from:

  - Vectorization

  - Efficient memory layout

  - Avoiding Python-level loops

  - Low-level C implementation

Understanding this topic means you understand why NumPy is fast.

# 1. Vectorization
## What Is Vectorization?

- Vectorization means performing operations on entire arrays at once, instead of looping over elements one by one.
- NumPy internally executes vectorized operations in optimized C code, not Python.

```python
# Example: Without Vectorization (Slow)
import numpy as np

a = np.arange(1_000_000)
b = np.arange(1_000_000)

c = []
for i in range(len(a)):
    c.append(a[i] + b[i])
```

**Slow because:**

 - Python loop overhead

 - Dynamic typing

 - Interpreter execution per iteration

**Example: With Vectorization (Fast)**
- No explicit loop
- Uses low-level optimized C routines
- Extremely fast

```python
c = a + b
```

```python
print(c)
```

## **Why Vectorization Is Faster**
| Reason             | Explanation                       |
| ------------------ | --------------------------------- |
| Fewer Python calls | One operation instead of millions |
| CPU cache friendly | Sequential memory access          |
| SIMD instructions  | Uses CPU vector units             |
| Optimized loops    | Written in C, not Python          |


# 2. Avoiding Python Loops
### Why Python Loops Are Slow
- Python loops are slow because:

  - Each iteration is interpreted

  - Dynamic type checking

  - Function call overhead

```python
# Example: Loop vs NumPy
# Python Loop
result = []
for x in a:
    result.append(x * 2)
```

```python
# NumPy Vectorized
result = a * 2
```

- Faster
- Cleaner
- Less memory overh

**Rule of Thumb**
- If you see a loop over NumPy arrays, there is probably a vectorized solution.

# 3. Speed Comparison (Practical Demonstration)
### Measuring Speed
- We can use timeit to compare execution time.

```python
# Python Loop
import time

start = time.time()
res = []
for x in range(1_000_000):
    res.append(x * 2)
end = time.time()

print("Loop time:", end - start)
```

```python
# NumPy Vectorized
import numpy as np
import time

arr = np.arange(1_000_000)

start = time.time()
res = arr * 2
end = time.time()

print("NumPy time:", end - start)
```

# 4. Strides & Memory Layout
### What Are Strides?
- Strides tell NumPy:
  - How many bytes to move in memory to get to the next element along each dimension.

- Understanding with Example

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]], dtype=np.int32)

print(arr.strides)
```

- Explanation
   - Each int32 takes 4 bytes
   - Row has 3 elements → 12 bytes
<br><br>
- strides = (12, 4)
   - Meaning:
      - Move 12 bytes to go to next row
      - Move 4 bytes to go to next column
