# What is Shape Manipulation?
- Shape manipulation means changing the structure (dimensions and layout) of a NumPy array without changing the actual data values.

- Example:

  - Convert 1D → 2D

  - Convert matrix → vector

  - Add or remove dimensions

- Why Shape Manipulation is Needed

  - ML models require specific input shapes

  - Data visualization needs reshaping

  - Broadcasting requires compatible shapes

  - Image data needs dimensional control

# 1. reshape()
   - reshape() changes the shape (dimensions) of an array without changing the data.
- Syntax
   - array.reshape(new_shape)
 
- Important Rule
   - Total number of elements MUST remain the same


```python
import numpy as np

a = np.array([1,2,3,4,5,6])

b = a.reshape(2,3) # (row, column)
print(b)
```

```python
# 2D to 3D Example
c = np.arange(12)
print(c.reshape(2,2,3)) #(array number, number of rows in per array, number of column in per array)
```

- reshape() returns a VIEW (if possible)

  - This means:

      - Data is shared

      - Changing one may affect the other

# 2. reshape(-1)
- -1 tells NumPy:
   - “Automatically calculate this dimension for me”

```python
a = np.array([[1,2,3],
              [4,5,6]])

print(a.reshape(-1))
```

- Use Cases

   - Convert multi-dimensional array into 1D

   - Prepare data for ML models

# 3. ravel()
- ravel() flattens a multi-dimensional array into 1D, returns a VIEW whenever possible.

- Syntax
   - array.ravel()

```python
a = np.array([[1,2],
              [3,4]])

b = a.ravel()
print(b)
```

- Key Property

   - Shares memory

   - Faster than flatten()

```python
b[0] = 100
print(a)
```

# 4. flatten()
- flatten() also converts array into 1D, but it always returns a COPY.

```python
arr1 = np.array([
    [
        [23,4,5],
        [6,7,4]
    ],

    [
        [6,7,4],
        [8,1,2]
    ]
])
```

```python
print(arr1.flatten())
```

```python
print(arr1)
```

```python
a = np.array([[1,2],
              [3,4]])

b = a.flatten()
b[0] = 100

print(a)
print(b)
```

- Difference from ravel()
| Feature           | ravel() | flatten() |
| ----------------- | ------- | --------- |
| Memory            | View    | Copy      |
| Speed             | Faster  | Slower    |
| Safe modification | ❌      | ✔         |


5. transpose()
- transpose() swaps rows and columns (axes).

- Syntax

   - array.transpose()

```python
# 2D Example
a = np.array([[1,2,3],
              [4,5,6]])

print(a.transpose())
```

6. T Attribute
- T is a shortcut for transpose.

```python
print(a.T)
```

- Important Note

   - Works same as transpose()

   - Very commonly used in ML

# 7. squeeze()
 - Removes dimensions of size 1.

```python
a = np.array([[[10, 20, 30]]])
print(a.shape)

b = np.squeeze(a)
print(b.shape)
```

- Use Case

  - Remove unnecessary dimensions

  - Clean ML input/output shapes

# 8. expand_dims()
- Adds a new axis at a specified position.

- Syntax
   - np.expand_dims(array, axis)

```python
a = np.array([1,2,3])

b = np.expand_dims(a, axis=0)
c = np.expand_dims(a, axis=1)

print(b.shape)
print(c.shape)
```

- Why Important

  - Required for broadcasting

  - Used before feeding data into ML models

- Complete Comparison Table
| Function      | Purpose            | View/Copy          |
| ------------- | ------------------ | ------------------ |
| reshape()     | Change shape       | View (if possible) |
| reshape(-1)   | Auto flatten       | View               |
| ravel()       | Flatten            | View               |
| flatten()     | Flatten            | Copy               |
| transpose()   | Swap axes          | View               |
| T             | Shortcut transpose | View               |
| squeeze()     | Remove axis of 1   | View               |
| expand_dims() | Add new axis       | View               |
