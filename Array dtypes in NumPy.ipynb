# What is dtype in NumPy?
- In NumPy, dtype (data type) defines what kind of data is stored in a NumPy array.

- It specifies:

  - Type of data (int, float, bool, etc.)

  - Size of each element in memory

  - How NumPy interprets the binary data

# Why dtype is Important

 - Controls memory usage

 - Determines precision of data

 - Affects speed of computation

 - Ensures data consistency

# Difference Between Python List and NumPy dtype
- Python List
   - Stores different data types

   - Each element is a separate Python object

   - More memory usage

```python
data = [1, 2.5, "hello"]
```

- NumPy Array
  - Stores same data type

  - Fixed memory block

  - Faster operations

```python
import numpy as np
arr = np.array([1, 2, 3])
```

# Checking dtype of an Array
  - Using dtype attribute

```python
arr = np.array([10, 20, 30])
print(arr.dtype)
```

# Common NumPy Data Types (dtypes)
1. Integer Types
| dtype | Description    |
| ----- | -------------- |
| int8  | 8-bit integer  |
| int16 | 16-bit integer |
| int32 | 32-bit integer |
| int64 | 64-bit integer |

```python
np.array([1,2,3], dtype=np.int32)
```

2. Floating-Point Types
| dtype   | Description                |
| ------- | -------------------------- |
| float16 | Half precision             |
| float32 | Single precision           |
| float64 | Double precision (default) |

```python
np.array([1.5, 2.3], dtype=np.float32)
```

3. Boolean Type
- Stores True or False.

```python
arr_b = np.array([True, False, True])
```

```python
print(arr_b.dtype)
```

4. Complex Numbers

```python
arr_c = np.array([1+2j, 3+4j])
print(arr_c.dtype)
```

5. String Type

```python
arr_s = np.array(["apple", "banana", "cherry"])
print(arr_s.dtype)
```

6. Object Type

- Used when data types are mixed.


```python
arr_o = np.array([1, "hello", 2.5], dtype=object)
print(arr_o.dtype)
```

# Automatic dtype Conversion (Type Promotion)

```python
arr_a = np.array([1, 2.5, 3])
print(arr_a)
print(arr_a.dtype)
```

- NumPy converts int ŌåÆ float
- Because float can store integer values

- Specifying dtype Manually

```python
arr = np.array([1,2,3], dtype=float)
```

- Memory Comparison

```python
a = np.array([1,2,3], dtype=np.int8)
b = np.array([1,2,3], dtype=np.int64)

print(a.itemsize)
print(b.itemsize)
```

# Changing dtype (astype())
- Used to convert array to another dtype.

```python
arr_d = np.array([1,2,3])
new_arr = arr_d.astype(float)

print(new_arr)
print(new_arr.dtype)
```

- Important Note

   - astype() creates a new copy

   - Original array remains unchanged

# itemsize and nbytes

- itemsize

   - Size of one element in bytes.

```python
arr = np.array([1,2,3], dtype=np.int32)
print(arr.itemsize)
```

- nbytes

  - Total memory used.

```python
print(arr.nbytes)
```

```python
arr1 = np.array([1.123456789], dtype=np.float32)
arr2 = np.array([1.123456789], dtype=np.float64)

print(arr1)
print(arr2)
```
