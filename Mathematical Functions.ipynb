# What are Mathematical Functions in NumPy?
- NumPy mathematical functions are built-in optimized functions that perform:

   - Statistical calculations

   - Mathematical transformations

   - Scientific operations , on NumPy arrays efficiently and accurately.

- These functions work in a vectorized manner, meaning they operate on entire arrays at once.

# 1. Statistical Functions

- 1.1 sum()
  - Calculates the total sum of all elements in an array.

- Syntax
  - np.sum(array, axis=None)

```python
import numpy as np
a = np.array([10, 20, 30, 40])
print(np.sum(a))
```

```python
# 2D Example with axis
b = np.array([[1,2,3],
              [4,5,6]])

print(np.sum(b, axis=0))  # column sum
print(np.sum(b, axis=1))  # row sum
```

- 1.2 mean()
  - Calculates the average value.

- Formula
   - mean = sum of values / total number of values

```python
print(np.mean(a))
```

- 1.3 median()
  - Returns the middle value after sorting the data.

- Rules

   - Odd count → middle element

   - Even count → average of two middle values

```python
c = np.array([10, 50, 20, 40, 30])

print(np.median(c))
```

- 1.4 std() – Standard Deviation
   - Measures how spread out the values are from the mean.

- Concept

   - Small std → values are close to mean

   - Large std → values are widely spread

```python
print(np.std(a))
```

- 1.5 var() – Variance
   - Variance is the square of standard deviation.

- Formula
   - variance = std²

```python
print(np.var(a))
```

# 2. Minimum & Maximum Functions

- 2.1 max() and min()
    - Returns the largest and smallest element.

```python
print(np.max(a))
print(np.min(a))
```

- 2.2 argmax() and argmin()
   - Returns the index position of the maximum or minimum value

```python
d = np.array([10, 25, 5, 40])

print(np.argmax(d))
print(np.argmin(d))
```

# 3. Power & Logarithmic Functions

- 3.1 sqrt()
  - Calculates square root of each element.

```python
x = np.array([4, 9, 16])

print(np.sqrt(x))
```

- 3.2 power()
  - Raises each element to a given power.

```python
print(np.power(x, 2))
```

- 3.3 exp()
   - Calculates e raised to power x.

```python
print(np.exp([1, 2, 3]))
```

- 3.4 log() (Natural Log)
  - Computes log base e.

```python
print(np.log([1, np.e, 10]))
```

- 3.5 log10()
  - Computes log base 10.

```python
print(np.log10([1, 10, 100]))
```

- Log functions do not accept zero or negative values.

# 4. Trigonometric Functions
- IMPORTANT RULE

   - NumPy trigonometric functions use radians, not degrees.

- Common Trigonometric Functions
| Function | Meaning         |
| -------- | --------------- |
| sin()    | Sine            |
| cos()    | Cosine          |
| tan()    | Tangent         |
| arcsin() | Inverse sine    |
| arccos() | Inverse cosine  |
| arctan() | Inverse tangent |

```python
angles = np.array([0, np.pi/2, np.pi])

print(np.sin(angles))
print(np.cos(angles))
```

```python
deg = np.array([0, 30, 90])
rad = np.deg2rad(deg)

print(np.sin(rad))
```

# 5. Rounding Functions

- 5.1 floor()
  - Rounds down to nearest integer.

```python
a = np.array([2.9, 3.1, -2.7])

print(np.floor(a))
```

- 5.2 ceil()
  - mRounds up to nearest integer.

```python
print(np.ceil(a))
```

- 5.3 round()
  - Rounds to nearest value.

```python
b = np.array([1.45, 1.55])

print(np.round(b))
```

```python
# Decimal Places
print(np.round([1.2345, 2.6789], 2))
```

- 5.4 truncate()
   - Removes decimal part without rounding.

```python
print(np.trunc(a))
```
