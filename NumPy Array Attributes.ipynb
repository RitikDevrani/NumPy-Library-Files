- Array initialization

```python
zero_array = np.zeros((3,4)) # all elements will be zero.
print(zero_array)
```

```python
print(np.zeros(5))
print(np.zeros((2,3)))
```

```python
ones_array = np.ones((5,4)) # all elements will be 1.
print(ones_array)
```

```python
print(np.ones(4))
print(np.ones((3,2)))
```

```python
full_arr = np.full((2,3),12) # full array: all elements will be same based on given values.
print(full_arr)
```

```python
id_arr = np.eye(4) # Identity matrix : Diagonal element values will b 1 and all other element's will be zero.
print(id_arr)
```

```python
print(np.empty(3)) # empty array : 
```

```python
print(np.arange(2,10)) # evenly spaced array: element are evenly spaced based on step values.
```

```python
print(np.arange(2,10,2))
```

```python
arrL = np.linspace(0,10,20) # specified number of equaly spaced values between a range.
arrL
```

```python
arr = np.random.rand(10) #random values array - float
print(arr)
```

```python
arr = np.random.rand(3,4)
arr
```

```python
random_int = np.random.randint(1,10) # random value - int
print(random_int)
```

```python
random_int = np.random.randint(1,10,4) # random values  based on given number(4) - int 
print(random_int)
```
