# What is a NumPy Array (ndarray)?
- A NumPy Array is a special data structure provided by NumPy, called ndarray (N-Dimensional Array), which is used to store multiple values of the same data type in a continuous block of memory.


- Meaning of "ndarray"

    - n → number of dimensions
    - d → dimensional
    - array → collection of elements


- So, ndarray means:
  - An array that can have 1 or more dimensions.

# Key Characteristics of NumPy Arrays
1.	Stores elements of same data type
2.	Faster than Python lists
3.	Supports multi-dimensional data
4.	Allows vectorized operations
5.	Uses less memory


```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print(arr)
print(type(arr))
```

# Difference Between Python List and NumPy Array
| Feature         | Python List   | NumPy Array |
| --------------- | ------------- | ----------- |
| Data Type       | Mixed allowed | Same only   |
| Speed           | Slow          | Fast        |
| Memory          | More          | Less        |
| Math Operations | Loop needed   | Direct      |
| Dimensions      | Limited       | Multiple    |


# Creating NumPy Arrays 

- There are three major ways to create NumPy arrays:

   - From Lists

   - From Tuples

   - Using Built-in Functions

# A) Creating NumPy Array from Lists

1. Dimensional Array from List
- A 1D array is a linear array containing elements in a single row.

```python
list_data = [1, 2, 3, 4, 5]
arr = np.array(list_data)

print(arr)
print(type(arr))
```

- Important 

  - Python list → converted into NumPy array

  - Data type is automatically detected

```python
#Check Data Type
print(arr.dtype)
```

2. Dimensional Array from List
- A 2D array is like a matrix (rows and columns).

```python
list_2d = [[1,2,3],
           [4,5,6],
           [7,8,9]]

arr2d = np.array(list_2d)

print(arr2d)
```

```python
#Check Dimensions
print(arr2d.ndim)    # Number of dimensions
print(arr2d.shape)  # Rows and columns
```

3. Mixed Data in List
- NumPy converts all elements to a common data type
- Result becomes float

# B) Creating NumPy Array from Tuples
- A tuple is similar to a list but immutable.
- NumPy can convert tuples into arrays easily.

```python
# 1D Array
import numpy as np

tuple_data = (10, 20, 30, 40)
arr = np.array(tuple_data)

print(arr)
```

```python
# 2D Array from Tuple
tuple_2d = ((1,2,3),
            (4,5,6))

arr2 = np.array(tuple_2d)
print(arr2)
```

# Why Use Tuples?
- Fixed data
- Slightly faster creation
- Useful in constant datasets

# C) Creating NumPy Arrays Using Built-in Functions
- These functions help create arrays without manually writing elements.

1. np.zeros()
- Creates an array filled with 0.

```python
arr = np.zeros(5)
print(arr)
```

```python
# 2D Zeros Array
np.zeros((3,4))
```

2. np.ones()
- Creates an array filled with 1.

```python
np.ones(4)
```

```python
# 2D Ones Array
np.ones((2,3))
```

3. np.full()
- Creates array filled with a specific value.

```python
np.full((3,3), 7)
```

4. np.empty()
- Creates array with random garbage values (uninitialized).

```python
np.empty(4)
# Values are unpredictabl
```

5. np.arange()
- Creates array with values in a range (similar to range()).
- Syntax
   - np.arange(start, stop, step)

```python
np.arange(1, 10, 2)
```

6. np.linspace()
- Generates evenly spaced numbers between a range.

```python
np.linspace(0, 1, 5)
```

7. np.eye() (Identity Matrix)
- Creates an identity matrix.

```python
np.eye(3)
```

- Important Observations

   - NumPy arrays store same type of data

   - Built-in functions save time

   - Arrays are memory efficient

   - Multi-dimensional arrays are easy
