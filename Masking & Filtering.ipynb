# Masking & Filtering in NumPy

- Masking and filtering allow you to select, modify, or remove elements from arrays based on conditions, without using loops.

- This concept is fundamental to:

   - Data cleaning

   - Feature selection

   - Conditional transformations

   - ML preprocessing

# What Is Masking?

- A mask is a Boolean array (True / False) that has the same shape as the original array.

   - True → keep / select the value

   - False → ignore / discard the value

- NumPy uses masks to perform fast, vectorized filtering.

# **1. Boolean Masks**
- How Boolean Masks Are Created

  - A Boolean mask is created using comparison operators:

| Operator | Meaning               |
| -------- | --------------------- |
| `>`      | greater than          |
| `<`      | less than             |
| `>=`     | greater than or equal |
| `<=`     | less than or equal    |
| `==`     | equal                 |
| `!=`     | not equal             |

```python
# Example: Creating a Boolean Mask
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

mask = arr > 30
print(mask)
```

- Each element is checked individually
- Result is a Boolean array

- **Using Mask to Filter Data**

```python
filtered = arr[arr > 30]
print(filtered)  #Only values where condition is True are selected
```

- **Why Boolean Masking Is Powerful**

   - No loops

   - Very fast

   - Easy to read

   - Memory efficient

# **2. Filtering Based on Conditions**

- Single Condition Filtering

```python
arr = np.array([5, 15, 25, 35, 45])

result = arr[arr >= 25]
print(result)
```

- **Multiple Conditions Filtering**

  - Use logical operators
| Operator | Meaning |    |
| -------- | ------- | -- |
| `&`      | AND     |    |
| `        | `       | OR |
| `~`      | NOT     |    |

- Always use parentheses.

```python
# Example: AND Condition
arr = np.array([10, 20, 30, 40, 50])

result = arr[(arr > 20) & (arr < 50)]
print(result)
```

```python
# Example: OR Condition
result = arr[(arr < 20) | (arr > 40)]
print(result)
```

```python
# Example: NOT Condition
result = arr[~(arr == 30)]
print(result)
```

# **Masking in 2D Arrays**

```python
# Example: Filter Values from Matrix
mat = np.array([[10, 25, 30],
                [40, 15, 50]])

mask = mat > 30
print(mask)
```

- Selecting Values

```python
print(mat[mat > 30])
```

- Modifying Values Using Mask

```python
mat[mat < 20] = 0
print(mat)

# Values less than 20 replaced with 0
```

# **3. np.where(condition, x, y)**

- np.where() works like a vectorized IF-ELSE statement.

  - If condition is True → take x  
  - If condition is False → take y

- Syntax
  - np.where(condition, x, y)

```python
# Example 1: Replace Values Conditionally
arr = np.array([10, 25, 30, 45])

result = np.where(arr > 30, 1, 0)
print(result)
```

```python
# Example 2: Conditional Transformation
arr = np.array([100, 200, 300])

result = np.where(arr > 150, arr * 2, arr / 2)
print(result)
```

```python
# Example 3: np.where() for Filtering Indices
arr = np.array([5, 15, 25, 35])

indices = np.where(arr > 20)
print(indices)
```

**Difference: Masking vs np.where()**
| Feature         | Boolean Mask | np.where |
| --------------- | ------------ | -------- |
| Filtering       | Yes          | No       |
| Replacement     | Yes          | Yes      |
| IF-ELSE logic   | Limited      | Full     |
| Returns indices | No           | Yes      |
