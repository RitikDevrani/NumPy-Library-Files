# Numpy
- Numpy is a library used for mathematical computations with multi-dimensional numerical data
  > Numpy is a low-level library written in C and FORTRAN for high-level mathematical functions. Numpy cleverly overcomes the problem of running slower algorithm in Python by using multidimensional arrays and functions that operate on arrays. Any algorithm can be expressed as a function on arrays, which allows the algorithms to be executed quickly and efficienly.

# What is an array?
- An array is a central data structure of the Numpy libray.
- It is a grid of homogeneous values (same data type).
- The rank of the array is the number of dimensions
- The array object in NumPy is called ndarray(n dimensional array).

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.dot(a, b)
print(result)
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(np.dot(A, B))
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[2, 0],
              [1, 2]])

print(A @ B)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.cross(a, b))
```

```python
A = np.array([[1, 2],
              [3, 4]])

print(np.linalg.det(A))
```

```python
A = np.array([[1, 2],
              [3, 4]])

A_inv = np.linalg.inv(A)
print(A_inv)
```

```python
print(A @ A_inv)
```

```python
A = np.array([[2, 1],
              [1, 3]])

B = np.array([5, 6])
X = np.linalg.solve(A, B)
print(X)
```
