# What is the NumPy Random Module?
- The NumPy random module is used to generate random numbers and random arrays for:

    - Simulations

    - Machine learning model initialization

    - Data sampling

    - Statistical experiments

    - Testing and debugging

 Random numbers generated by NumPy are pseudo-random, meaning they are generated by mathematical algorithms.

## Why Random Numbers Are Important

    - ML algorithms start with random weights

    - Train-test split requires randomness

    - Monte Carlo simulations

    - Random sampling of data

### 1. np.random.rand()
- Generates random numbers from a uniform distribution in the range [0, 1).

- Syntax
   - np.random.rand(d0, d1, d2, ...)


Each argument represents a dimension size.

- Key Properties

    - Values are ≥ 0 and < 1

    - Uniform distribution

    - Output is a NumPy array

```python
# Example: 1D Array
import numpy as np

a = np.random.rand(5)
print(a)
```

```python
# Example: 2D Array
b = np.random.rand(3, 4)
print(b)
```

```python
# Example: 3D Array
c = np.random.rand(2, 2, 3)
print(c)
```

### 2. np.random.randn()
- Generates random numbers from a standard normal distribution
- (mean = 0, standard deviation = 1).

    - Difference from rand()
| rand()               | randn()                     |
| -------------------- | --------------------------- |
| Uniform distribution | Normal distribution         |
| Range: [0,1)         | Range: (−∞, +∞)             |
| No negative values   | Can produce negative values |

```python
a = np.random.randn(5)
print(a)
```

```python
# 2D Example
b = np.random.randn(3, 3)
print(b)
```

### 3. np.random.randint()
- Generates random integers within a specified range.

- Syntax
   - np.random.randint(low, high=None, size=None)

- Important Rules

   - low is inclusive

   - high is exclusive

   - Can generate single value or array

- Single Integer

```python
print(np.random.randint(1, 10))
```

```python
# 1D Array
a = np.random.randint(10, 50, size=5)
print(a)
```

```python
# 2D Matrix
b = np.random.randint(0, 100, size=(3, 4))
print(b)
```

### 4. np.random.random()
- Generates random numbers from uniform distribution [0,1).

   - Difference from rand()
| random()            | rand()                     |
| ------------------- | -------------------------- |
| Accepts tuple shape | Accepts separate arguments |


```python
a = np.random.random(5)
print(a)
```

```python
# 2D Example
b = np.random.random((3, 3))
print(b)
```

### 5. np.random.choice()
- Randomly selects elements from a given array or range.

- Syntax
    - np.random.choice(a, size=None, replace=True, p=None)

- Parameters Explained

    - a → input array or integer range

    - size → output shape

    - replace → allow repetition or not

    - p → probability distribution

- Example: Choose from array

```python
a = np.array([10, 20, 30, 40])

print(np.random.choice(a))
```

- Choose multiple values

```python
print(np.random.choice(a, size=3))
```

- Without replacement

```python
print(np.random.choice(a, size=3, replace=False))
```

- With probabilities

```python
print(np.random.choice(a, size=5, p=[0.1, 0.2, 0.3, 0.4]))
```

### 6. np.random.seed()
- Sets the seed value to make random numbers reproducible.

- Why It Is Important

   - Debugging

   - Research experiments

   - Model reproducibility

```python
np.random.seed(42)
print(np.random.rand(3))

np.random.seed(42)
print(np.random.rand(3))
```

- Output will be exactly the same.

### 7. Creating Random Matrices

- Uniform Random Matrix

```python
matrix = np.random.rand(3, 4)
print(matrix)
```

- Normal Distribution Matrix

```python
matrix = np.random.randn(3, 4)
print(matrix)
```

- Integer Random Matrix

```python
matrix = np.random.randint(1, 100, size=(3, 4))
print(matrix)
```

- Binary Random Matrix

```python
matrix = np.random.randint(0, 2, size=(4, 4))
print(matrix)
```

- Probability-Based Matrix

```python
matrix = np.random.choice([0,1], size=(3,3), p=[0.7, 0.3])
print(matrix)
```
