# Statistical Operations in NumPy

- Statistical operations are used to analyze numerical data and extract meaningful information such as average, spread, variation, and distribution.

- These operations are extremely important in:

   - Data Analysis

   - Data Science

   - Machine Learning

   - Research & Engineering

### 1. mean()
 - The mean is the average of all elements.

- NumPy Function
   - np.mean(array)

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
print(np.mean(arr))
```

```python
# Axis Usage
arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])

print(np.mean(arr2, axis=0))  # column-wise
print(np.mean(arr2, axis=1))  # row-wise
```

### 2. median()
- The median is the middle value after sorting the data.

   - Odd count → middle value

   - Even count → average of two middle values

- NumPy Function
   - np.median(array)

```python
arr = np.array([5, 1, 9, 3, 7])
print(np.median(arr))
```

- Sorted: [1, 3, 5, 7, 9]
- Median = 5

### 3. std() – Standard Deviation
- Standard deviation measures how much data deviates from the mean.

   - Low std → data close to mean

   - High std → data spread out

- NumPy Function
   - np.std(array)

```python
arr = np.array([10, 12, 14, 16, 18])
print(np.std(arr))
```

### 4. var() – Variance
- Variance is the square of standard deviation.

- Variance =(Standard Deviation)**2

- NumPy Function
   - np.var(array)

```python
arr = np.array([10, 12, 14, 16, 18])
print(np.var(arr))
```

### 5. ptp() – Peak to Peak
- Returns the difference between maximum and minimum value.
- ptp = max − min

- NumPy Function
   - np.ptp(array)

```python
arr = np.array([5, 10, 20, 40])
print(np.ptp(arr))
```

### 6. percentile()
- A percentile tells the value below which a given percentage of data lies.

- Examples:

  - 25th percentile → Q1

  - 50th percentile → Median

  - 75th percentile → Q3

- NumPy Function
  - np.percentile(array, percentile_value)

```python
arr = np.array([10, 20, 30, 40, 50])
print(np.percentile(arr, 25))
print(np.percentile(arr, 50))
print(np.percentile(arr, 75))
```

### 7. histogram()
- A histogram shows frequency distribution of data.

- NumPy Function
   - np.histogram(array, bins)

```python
arr = np.array([1, 2, 2, 3, 3, 3, 4, 5])

hist, bins = np.histogram(arr, bins=4)
print(hist)
print(bins)
```

- Output

  - hist → frequency

  - bins → data range intervals
